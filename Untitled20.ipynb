```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from numpy import std
seed(1)
data = pd.read_excel('/content/drive/MyDrive/кз2-2/Книга1.xlsx')
data1 = data[['Gross (A$m)']]
print('mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
#data
```

```python
from matplotlib import pyplot
seed(1)
pyplot.hist(data1)
pyplot.show()
```

```python
from statsmodels.graphics.gofplots import qqplot
seed(1)
qqplot(data1, line='s')
pyplot.show()
```

```python
from scipy.stats import shapiro
seed(1)
stat, p=shapiro(data1)
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
  print('Sample looks Gaussian (fail to reject HO)')
else:
  print('Sample does not look Gaussian (reject HO)')
```

```python
from scipy.stats import normaltest
seed(1)
stat, p = normaltest(data1)
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
  print('Sample looks Gaussian (fail to reject HO)')
else:
  print('Sample does not look Gaussian (reject HO)')
```

```python
from scipy.stats import anderson
seed(1)
result = anderson(data1.values.flatten())
print('Statistics=%.3f' % result.statistic)
p=0
for i in range(len(result.critical_values)):
  s1, cv = result.significance_level[i], result.critical_values[i]
  if result.statistic < result.critical_values[i]:
    print('%.3f: %.3f, data1 looks normal (fail to reject HO)' % (s1,cv))
  else:
    print('%.3f: %.3f, data1 does not look normal (reject HO)' % (s1,cv))
```
