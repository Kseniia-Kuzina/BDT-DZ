```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/titanic.csv')
```

```python
data
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Sex', y='Age', data=data)
plt.title('Box plot')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Survived', y='Age', data=data)
plt.title('Violin plot of Age by Survived')
plt.show()
```

```python
# Count plots
sns.countplot(x='Sex', data=data)
```

```python
# Heatmap
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
```

```python
# Facet grid
g = sns.FacetGrid(data, col='Sex', row='Age')
g.map(sns.scatterplot, 'Pclass', 'Survived')
```

```python
# Regplot
sns.regplot(x='Pclass', y='Survived', data=data)
```

```python
# lmplot
sns.lmplot(x='Pclass', y='Survived', data=data, hue='Sex')
```

```python
# Swarm plot
sns.swarmplot(x='Sex', y='Age', data=data)
```

```python
# Pair plot
sns.pairplot(data, hue='Sex')
```

```python
# Joint plot
sns.jointplot(x='Age', y='Pclass', data=data, kind='hex')
```
