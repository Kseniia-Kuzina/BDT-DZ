```
#   5.1. Создайте из таблицы  DataFrame.

tabl = [[51.0,70.0,64.0,81.0],[52.0,70.5,64.2,81.3],[51.4,69.1,66.8,80.5],[52.8,69.8,66.0,80.9],[50.5,70.5,63.4,81.4]]
df = pd.DataFrame(tabl,columns=['Anna', 'Ben','Carole','Dave'])
df
```

```
#  5.2. Преобразуйте этот DataFrame в tidy format.

tidy_df = df.reset_index().melt(id_vars='index', value_vars=['Anna', 'Ben', 'Carole', 'Dave'],
                                var_name='Name', value_name='weight')
tidy_df
```

```
#  5.3. Определите, у кого из этих четырех человек было меньше всего колебаний по весу.
min_std_dev_person = tidy_df.groupby('Name')['weight'].std().idxmin()
min_std_dev_person
# ваш код
#df_melted.groupby('name').agg({'weight':np.std})
```
