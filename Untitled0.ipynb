```python
from pandas import read_excel
from sklearn.preprocessing import MinMaxScaler
series = read_excel ('/content/drive/MyDrive/кз 2-3/проба.xlsx', header=0, index_col=0)
print(series.head())

values = series.values
values = values[:1000]
values = values.reshape(len(values), 7)

scaler = MinMaxScaler(feature_range = (0,1))
scaler.fit(values)
print('Min: %f, Max: %f' % (scaler.data_min_.flatten()[0], scaler.data_max_.flatten()[0]))

normalized = scaler.transform(values)
for i in range(5):
  print(normalized[i])

inversed = scaler.inverse_transform(normalized)
for i in range(5):
  print(inversed[i])
```

```python
from pandas import read_excel
from sklearn.preprocessing import StandardScaler
from math import sqrt

series = read_excel ('/content/drive/MyDrive/кз 2-3/проба.xlsx', header=0, index_col=0)
print(series.head())

values = series.values
desired_shape = (7000, 1)
values = values[:1000]
values = values.reshape(desired_shape)

scaler = StandardScaler()
scaler = scaler.fit(values)
print('Mean: %f, StandardDeviation: %f' % (scaler.mean_, sqrt(scaler.var_)))

normalized = scaler.transform(values)
for i in range(5):
  print(normalized[i])

inversed = scaler.inverse_transform(normalized)
for i in range(5):
  print(inversed[i])
```
